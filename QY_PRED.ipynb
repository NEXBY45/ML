```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.metrics import mean_squared_error
```

Loading the database.

```python
from google.colab.data_table import DataTable
DataTable.max_columns = 30
train_df = pd.read_csv('train.csv', dtype={'solvent':object,'ligands':object,'DOI':object})
```

```python
train_df
```

```python
train_df.describe()
```

Defining what is a variable and a function.

```python
x = train_df[['Pb_mmol','Yb_mmol', 'OlAm_mL','OlAc_mL','temperature_°C','time_sec']]
y = train_df['NIR QY_%']
```

```python
x.round(3)
```

```python
variables =pd.DataFrame({'Chemical formula':train_df['chem formula'],  'DOI':train_df['DOI']})
variables
```

```python
variables.join(x)
```

Train on an initial data using linear regression.

```python
from sklearn.linear_model import LinearRegression
from sklearn import linear_model
regr = linear_model.LinearRegression(fit_intercept = False)
regr.fit(x, y)
```

Let's make a prediction for all x and get the predicted values (column of predictions).

```python
pred=regr.predict(x)
compare =pd.DataFrame({'Prediction of NIR QY_%':pred, 'Experimental NIR QY_%':train_df['NIR QY_%'], 'Absolute difference_%': abs(pred-y)})
```

```python
pd.options.display.float_format = '{:,.1f}'.format
variables.join(compare)
```

Consider the resulting metrics: the Mean Squared Error (MSE), mean absolute error (MAE), Root Mean Squared Error (RMSE):

```python
from sklearn.metrics import mean_squared_error


mse = mean_squared_error(y, pred)
mae=abs(pred-y).mean()

rmse = mean_squared_error(train_df['NIR QY_%'], pred,squared=False)

print(f'MSE {mse:.2f}')
print(f'RMSE {rmse:.2f}')
print(f'MAE {mae:.2f}')
```

The coefficient of determination (R²) measures how well a statistical model predicts an outcome. The result is represented by the dependent variable of the model. The smallest possible value of R² is 0, and the maximum possible value is 1. Simply put, the better the model makes predictions, the closer its R² will be to 1.

```python
r_sq = regr.score(x, y)
print('coefficient of determination:', r_sq)
```

The intercept value (free weight) shows that your model predicts the value of the function when the variables in x are all zero.

```python
print('intercept:', regr.intercept_)
```

Let's derive what coefficients (feature weights) are obtained for the resulting machine learning model.

```python
print(regr.coef_)
```

```python
pd.options.display.float_format = '{:,.2f}'.format
finish =pd.DataFrame({'Variables':x.columns , 'Coefficients':regr.coef_})
finish
```

Let's set random parameters for the machine learning model so that it outputs the "predicted" value of the function:

```python
pred_QY = regr.predict([[0.55,0.55,1.6,1.6,240,60]])
print(pred_QY)
```
